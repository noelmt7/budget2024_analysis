```python
import pandas as pd
import numpy as np
from faker import Faker

fake = Faker()

Faker.seed(14)
np.random.seed(45)

num_rows = 100000

# Define possible opinions and sentiments
opinions = [
    "I believe the budget will greatly benefit the economy.",
    "The budget seems to favor the wealthy.",
    "It's a well-balanced budget.",
    "I'm concerned about the lack of support for small businesses.",
    "This budget does not address the needs of the common man.",
    "The healthcare spending is a step in the right direction.",
    "The tax reforms are long overdue.",
    "I'm not sure if this budget will really help the poor.",
    "This budget is a positive step towards economic growth.",
    "The education sector has been neglected."
]

sentiments = ['Positive', 'Negative', 'Neutral']

# Generate the data
data = pd.DataFrame({
    'opinion_id': [fake.uuid4() for _ in range(num_rows)],
    'name': [fake.name() for _ in range(num_rows)],
    'age': np.random.randint(18, 90, num_rows),
    'gender': [fake.random_element(elements=('Male', 'Female')) for _ in range(num_rows)],
    'location': [fake.city() for _ in range(num_rows)],
    'opinion': [fake.random_element(elements=opinions) for _ in range(num_rows)],
    'sentiment': [fake.random_element(elements=sentiments) for _ in range(num_rows)]
})

# Save to CSV
data.to_csv('consumer_opinions_budget2024.csv', index=False)

# Display the first few rows of the data
print(data.head())
```

```python
import pandas as pd
import numpy as np
from faker import Faker

fake = Faker()

Faker.seed(14)
np.random.seed(45)

num_rows = 100000

# Define possible opinions and sentiments for producers
opinions = [
    "The budget will positively impact the manufacturing sector.",
    "There is insufficient support for agricultural producers.",
    "The tax incentives will encourage more production.",
    "The budget does not address the rising costs of raw materials.",
    "We need more subsidies to remain competitive.",
    "The infrastructure spending will help our logistics.",
    "I'm worried about the increased compliance costs.",
    "The budget is a good step for the export industry.",
    "This budget will help us scale our production.",
    "The lack of focus on renewable energy is disappointing."
]

sentiments = ['Positive', 'Negative', 'Neutral']

# Generate the data
data = pd.DataFrame({
    'opinion_id': [fake.uuid4() for _ in range(num_rows)],
    'name': [fake.name() for _ in range(num_rows)],
    'age': np.random.randint(18, 90, num_rows),
    'gender': [fake.random_element(elements=('Male', 'Female')) for _ in range(num_rows)],
    'location': [fake.city() for _ in range(num_rows)],
    'opinion': [fake.random_element(elements=opinions) for _ in range(num_rows)],
    'sentiment': [fake.random_element(elements=sentiments) for _ in range(num_rows)]
})

# Save to CSV
data.to_csv('producer_opinions_budget2024.csv', index=False)

# Display the first few rows of the data
data.head()
```
